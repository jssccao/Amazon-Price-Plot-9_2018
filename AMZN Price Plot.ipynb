```python
import pandas as pd 
import matplotlib.pyplot as plt 

file = pd.read_csv('AMZN.csv', delimiter=',')
print(file)

plt.figure(figsize=(10,4))
ax = plt.subplot()
xlabels = []
for num in file['Date']:
    str(num)
    num = num[:-3]
    xlabels.append(num)

plt.scatter(file['Date'], file['Low'], marker='o')
plt.scatter(file['Date'], file['High'], marker='o')
plt.plot(file['Date'], file['Close'], color='green')

plt.title('AMZN September 2018 Prices')
plt.xlabel('Dates')
ax.set_xticklabels(xlabels, rotation=30)
plt.ylabel('Prices, $')
plt.legend(['Close','Low','High'], loc=4)

plt.show()
```
